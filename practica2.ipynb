```python
import sklearn.datasets as datasets
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data =  datasets.load_iris()
X = data.data[:, :2]  # we only take the first two features.
y = data.target
```

```python
Iris3D = PCA(n_components=3).fit_transform(data.data)
Iris2D = PCA(n_components=2).fit_transform(data.data)
```

```python
kmeansOriginal = KMeans(n_clusters=3, random_state=0).fit(X)
kmeans3D = KMeans(n_clusters=3, random_state=0).fit(Iris3D)
```

```python
kmeans2D = KMeans(n_clusters=3).fit(Iris2D)
```

```python
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)


index_selector3D = [np.unique(kmeans3D.labels_)[i] == kmeans3D.labels_ for i in range(len(np.unique(kmeans3D.labels_)))]
for i in range(len(index_selector3D)):
    ax.scatter(Iris3D[index_selector3D[i]].T[0],Iris3D[index_selector3D[i]].T[1],Iris3D[index_selector3D[i]].T[2])

ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])
```
